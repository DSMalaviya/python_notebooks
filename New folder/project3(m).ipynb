```
import cv2
```

```
#now convert all rgb to gray scale
```

```

files=os.listdir('normal')
for f in files:
  image = cv2.imread(os.path.join('normal',f))
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  cv2.imwrite(os.path.join('normal2',f),gray)
```
