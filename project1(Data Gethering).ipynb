```
#now the downloaded dataset have mixed type of imgs so I separate covid imgs from it
import pandas as pd
import shutil
import os
```

```
#change the name of github file to ds
#after it we make the target directory by below python commands
os.mkdir("Dataset")
```

```
os.mkdir("Dataset/covid")
```

```
#now csv files contains information about the imgs so by using the csv file I can eassily separate those imgs
Files_path='ds/metadata.csv'
Imgs_path='ds/images'
```

```
df=pd.read_csv(Files_path)
print(df.shape)
```

```
#so it contains total 930 imgs
df.head()
#for just viewing purpose
```

```
#now create target dir
Target_dir="Dataset/covid"
if not os.path.exists(Target_dir):
  os.mkdir(target_dir)
  print("folder created of covid data")

#above script dir nai hoi to create kari dese  
```

```
#lets find how many covid imgs are in dataset
cnt=0
for (i,row) in df.iterrows():
  if row['finding']=='COVID-19' and row['view']=="PA":
    cnt+=1

print(cnt)
```

```
#now we copy the imgs to our dataset folder but here not that we take only PA view of xray
cnt=0
for (i,row) in df.iterrows():
  if row['finding']=='COVID-19' and row['view']=="PA":
    filename=row['filename']
    image_path=os.path.join(Imgs_path,filename)
    image_copy_path=os.path.join(Target_dir,filename)
    shutil.copy2(image_path,image_copy_path)
    print("moving img : ",cnt)
    cnt+=1
```

```
#now covid dataset is ready and now we find normal dataset from kaggle
#first upload the json file
import os
os.environ['KAGGLE_CONFIG_DIR'] = "/content"
```

```
#now extract to downloaded zip file using below script
from zipfile import ZipFile
fname='chest-xray-pneumonia.zip'
with ZipFile(fname,'r') as zip:
  zip.extractall()
  print('done')
```

```
#after extracting I only take imgs from the train folder
#there are many imgs so I take only 180 imgs randomly from the train folder
# create normal folder in Dataset directory
os.mkdir('Dataset/normal')
```

```
Kaggle_file_path='chest_xray/train/NORMAL'
Target_path='Dataset/normal'
```

```
image_names=os.listdir(Kaggle_file_path)
```

```
import random
random.shuffle(image_names)
#above script will randomly shuffle all the imgs
```

```
cnt=1
for i in range(180):
  image_name=image_names[i]
  image_path=os.path.join(Kaggle_file_path,image_name)
  target_path=os.path.join(Target_path,image_name)
  shutil.copy2(image_path,target_path)
  cnt+=1
  print("coping img:",cnt)
```

```
#now our Dataset preparation is ready so I save to the drive so i will latter use it
#we make the zip file ane copy to our drive folder so it can be used after the runtime is disconnected
shutil.make_archive('Dataset', 'zip', 'Dataset')
```
